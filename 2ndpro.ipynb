```python
#merged_news_df['text'] = merged_news_df['text'].apply(stemming)
import re
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords

# Initialize the Porter Stemmer and download the stopwords if necessary
ps = PorterStemmer()
stop_words = set(stopwords.words('english'))

def stemming(text):
    # Remove non-alphabetical characters and convert to lowercase
    stemmed_text = re.sub('[^a-zA-Z]', ' ', text)
    stemmed_text = stemmed_text.lower()
    
    # Tokenize the text
    words = stemmed_text.split()
    
    # Apply stemming and remove stopwords
    stemmed_words = [ps.stem(word) for word in words if word not in stop_words]
    
    # Join the stemmed words back into a single string
    stemmed_text = ' '.join(stemmed_words)
    return stemmed_text

# Assuming 'merged_news_df' is your DataFrame, apply the 'stemming' function to the 'text' column
merged_news_df['text'] = merged_news_df['text'].apply(stemming)
```

```python
X = merged_news_df ['text'].values
y = merged_news_df ['Label'].values
```

```python
vector = TfidfVectorizer()
vector.fit(X)
X = vector.transform(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.2, stratify=y, random_state=2)
```

```python
model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
# on training set
train_y_pred = model.predict(X_train)
print(accuracy_score(train_y_pred,Y_train))
```

```python
# on testing set
testing_y_pred = model.predict(X_test)
print(accuracy_score(testing_y_pred,Y_test))
```
